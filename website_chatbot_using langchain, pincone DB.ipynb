```python
from langchain_pinecone import PineconeVectorStore
from pinecone import Pinecone, ServerlessSpec

pc = Pinecone(
        api_key=os.environ.get("PINECONE_API_KEY")
    )

index_name = "chatproject"

pc.create_index(
    name=index_name,
    dimension=768, # Replace with your model dimensions
    metric="cosine", # Replace with your model metric
    spec=ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    )
)
```

```python
vectorstore = PineconeVectorStore.from_texts(
    texts=[t.page_content for t in text_chunks],
    embedding=embedding,
    index_name=index_name
)
```

```python
notebook_login()
```

```python
model = "Qwen/Qwen3-8B"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model, use_auth_token=True)

model = AutoModelForCausalLM.from_pretrained(model,
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                             load_in_8bit=True)
```

```python
pipe = pipeline('text-generation',
                model=model,
                tokenizer=tokenizer,
                torch_dtype=torch.bfloat16,
                device_map='auto',
                max_new_tokens=512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm = HuggingFacePipeline(pipeline=pipe, model_kwargs={'temparature':0})
```

```python
llm.invoke('please provide a concise summary of the book harry potter')
```

```python
from langchain.chains import RetrievalQA
```

```python
query = "How much nvidia Total assists"
```

```python
docs = vectorstore.similarity_search(query, k=3)
```

```python
docs
```

```python
chain = RetrievalQA.from_chain_type(llm=llm,
                                    chain_type='stuff',
                                    retriever=vectorstore.as_retriever())
```

```python
query = "How much nvidia Total assists"
chain.run(query)
```

```python
query = "who are the key employees in OpenAI"
chain.run(query)
```

```python
while True:
  user_input = input(f'Input Prompt: ')
  if user_input == 'exist':
    print('Exiting')
    sys.exit()
  if user_input == '':
    continue
  result = chain({"query" : user_input})
  print(f'Answer: {result["result"]}')
```
